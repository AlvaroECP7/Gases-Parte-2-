```python
df = pd.read_csv("data (2).csv")
df
```

```python
df['y_inversa'] = 1 / df['y']

coefficients = np.polyfit(df['x'], df['y_inversa'], 1)
linea_tendencia_inversa = np.poly1d(coefficients)

plt.plot(df['x'], df["y"], label='Datos', marker='o', linestyle='None', color='green')

x_tendencia_inversa = np.linspace(df['x'].min(), df['x'].max(), 100)
y_tendencia_inversa = linea_tendencia_inversa(x_tendencia_inversa)
plt.plot(x_tendencia_inversa, y_tendencia_inversa, label='RegresiÃ³n lineal', color='blue', linewidth=4.0)


plt.xlabel('Volumen (litros)')
plt.ylabel('PresiÃ³n (atm)')
plt.title('GrÃ¡fica Ley de Boyle')

plt.grid(True)
plt.legend()

plt.show()
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://es.wikipedia.org/wiki/Gases_nobles"
response = requests.get(url)

if response.status_code == 200:

    soup = BeautifulSoup(response.text, 'html.parser')

    tabla = soup.find('table', {'class': 'wikitable'})

    df = pd.read_html(str(tabla))[0]

    df = df.rename(columns=df.iloc[0]).drop(df.index[0])

    print(df.head())

else:
    print("No se pudo obtener la pÃ¡gina web.")
```
